# Project Overview

* The goal of this project was to predict rental price growth across VIC suburbs over next 5 years,

* We did this by combining property-specific attributes (bedrooms, land size, car parks etc.) with external factors such as demographics (age), income, schools, and transport access


# Data

* Historical Rental Data
    * Quarterly median rental data for each suburb in Victoria (June 1999 - March 2025)

* School Data
    * Number of primary schools per suburb
    * Number of secondary schools per suburb
    * Proximity of rental properties to schools (using OpenRoute)

* Public Transport Data
    * Proximity of rental properties to metro buses, trains and trams

* Property-Level Data
    * Number of bedrooms
    * Number of bathrooms
    * Number of carspaces
    * etc.
