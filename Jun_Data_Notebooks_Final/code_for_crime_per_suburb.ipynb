```python
import pandas as pd
```

```python
df = pd.read_csv('data\external\Data_Tables_LGA_Recorded_Offences_Year_Ending_June_2025.csv')
print(len(df))
```

```python
crime_agg = df.groupby('Suburb/Town Name')['Offence Count'].sum().reset_index()
crime_agg['Suburb/Town Name'] = crime_agg['Suburb/Town Name'].str.lower()
```

```python
crime_agg
```

```python
crime_agg.to_csv('crime_per_suburb.csv', index=False)
```
